# Real Code

```python
for i in range(1, 3):
    for qtr in range(1, 5):
        num = 1
        for ch in pd.read_csv(f"local datasets/CSVs/202{i}Q{qtr}.csv", low_memory = False, sep = "|", header = None, usecols = imp, chunksize = sizeOfChunk):
            dfPart = ch.copy()
            dfPart.columns = header
            dfPart = dfPart[(dfPart['Loan Purpose'] == 'P')
                           & (dfPart['Property Type'] == 'SF')
                           & (dfPart['Amortization Type'] == 'FRM')
                           & (dfPart['Interest Only Loan Indicator'] == 'N') 
                           & (dfPart['Current Loan Delinquency Status'] == 0) 
                           & (dfPart['High Loan to Value (HLTV) Refinance Option Indicator'] == 'N')]
            dfPart = dfPart.drop(['Loan Purpose', 'Property Type', 'Amortization Type', 'Interest Only Loan Indicator', 'Current Loan Delinquency Status', 'High Loan to Value (HLTV) Refinance Option Indicator'], axis = 1)
            dfPart.to_csv(f"datasets/202{i}/Q{qtr}/202{i}Q{qtr}_P{num}.csv", index = False)
            num += 1
```
