## Storing unique loan id

```python
sizeOfChunk = int(1e6)
```

```python
sizeOfChunk
```

```python
unique_id = 0
```

```python
for i in range(19, 23):
    for qtr in range(1, 5):
        num = 1
        for ch in pd.read_csv(f"local datasets/CSVs/20{i}Q{qtr}.csv", low_memory = False, sep = "|", header = None, usecols = leftList, chunksize = sizeOfChunk):
            dfPart = ch.copy()
            dfPart.columns = header
            dfPart = dfPart[(dfPart['Loan Purpose'] == 'P')
                           & (dfPart['Property Type'] == 'SF')
                           & (dfPart['Amortization Type'] == 'FRM')
                           & (dfPart['Interest Only Loan Indicator'] == 'N') 
                           & (dfPart['High Loan to Value (HLTV) Refinance Option Indicator'] == 'N') 
                           & (dfPart['Original Loan Term'] == 360)]
            
            dfPart = dfPart.drop(['Loan Purpose', 'Property Type', 'Amortization Type', 'Interest Only Loan Indicator', 'High Loan to Value (HLTV) Refinance Option Indicator', 'Original Loan Term'], axis = 1)
            dfPart.to_csv(f"datasets/202{i}Q{qtr}_P{num}.csv", index = False)
            unique_id += dfPart['Loan Identifier'].nunique()
            break
```
